```python
import numpy as np
import matplotlib.pyplot as plt

import qopy.phase_space.wigner as wig
import qopy.utils.majorization as mjz
import qopy.plotting as wplt

import matplotlib as mpl
mpl.rcParams['text.usetex'] = True

rl = 15
nr = 1000

w0 = wig.fock(0, rl, nr)

fig, ax = plt.subplots(figsize=(12, 8))
states = range(0, 11)
colors = mpl.cm.viridis(np.linspace(0, 1, len(states)))  # option couleur auto

for n, c in zip(states, colors):
    p, q = mjz.relative_lorenz_decreasing_2d(wig.fock(n, rl, nr), w0, rl)
    ax.plot(q, p, label=rf'$|{n}\rangle$', color=c)

ax.legend(loc='lower right')
ax.set_xlabel('Cumulated integral of $|0\\rangle$')
ax.set_ylabel('Cumulated integral of $|n\\rangle$')
ax.set_title('Relative Lorenz curves of Fock states, relatively to vacuum')
fig.tight_layout()
plt.show()
```
