```python
import numpy as np

import qopy.phase_space.wigner as wig
import qopy.plotting as wplot
```

```python
rl = 20
nr = 1000


N = 3
amplitude_list = np.ones(N)
alpha_list = 5*np.exp(2*np.pi*1j*np.arange(N)/N)
xi_list = 0.5*np.exp(4*np.pi*1j*np.arange(N)/N)
n_list = np.ones(N, dtype=int)
w3 = wig.superposition.gaussian(alpha_list, xi_list, amplitude_list, rl, nr)

N = 4
amplitude_list = np.ones(N)
alpha_list = 5*np.exp(2*np.pi*1j*np.arange(N)/N)
xi_list = 0.5*np.exp(4*np.pi*1j*np.arange(N)/N)
n_list = np.ones(N, dtype=int)
w4 = wig.superposition.gaussian(alpha_list, xi_list, amplitude_list, rl, nr)

N = 5
amplitude_list = np.ones(N)
alpha_list = 5*np.exp(2*np.pi*1j*np.arange(N)/N)
xi_list = 0.5*np.exp(4*np.pi*1j*np.arange(N)/N)
n_list = np.ones(N, dtype=int)
w5 = wig.superposition.gaussian(alpha_list, xi_list, amplitude_list, rl, nr)

wplot.grid_2d([w3, w4, w5], rl)
```

```python
rl = 20
nr = 1000

N = 2
amplitude_list = np.ones(N)
alpha_list = [2, -2]
xi_list = [-1, -1]
n_list = np.ones(N, dtype=int)

w00 = wig.superposition.gaussian_fock(alpha_list, xi_list, [0, 0], amplitude_list, rl, nr)
w01 = wig.superposition.gaussian_fock(alpha_list, xi_list, [0, 1], amplitude_list, rl, nr)
w02 = wig.superposition.gaussian_fock(alpha_list, xi_list, [0, 2], amplitude_list, rl, nr)

wplot.grid_2d([w00, w01, w02], rl)
```
